```python
#Opening poc dataframe with AirStation 16 data
#############################################

poc_16 = pd.read_csv ("../data/processed/poc_16.csv", sep= ',')
poc_16.index = poc_16.TIME
poc_16 = poc_16[["NO2_index"]]
poc_16.head()
```

```python
#Splitting train and test dataset
#################################

train_data = poc_16['2019-01-01':'2020-08-01']
test_data = poc_16['2020-08-01':'2021-08-31']

print('Observations: %d' % (len(poc_16)))
print('Training Observations: %d' % (len(train_data)))
print('Testing Observations: %d' % (len(test_data)))
```

# AUTOARIMA

AutoARIMA function allow us to search the between the multiple combinations of the ARIMA parameters (p,d,q) and choose the best combination that has the least AIC, so the best model

```python
#AUTOARIMA EVALUATION 
######################



model = pm.auto_arima(train_data.NO2_index, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)


print(model.summary())
```

```python
#Plotting diagnostics graphs to evaluate the result
##################################################

model.plot_diagnostics(figsize=(20,14))
plt.show()
```

**DIAGNOSTICs PLOT ANALYSIS**

STANDARDIZED RESIDUAL: Uniform varianc around mean of 0.

HISTOGRAM ESTIMATED DENSITY: Distribution looks like to be Normal Distribution (with mean 0)

NORMAL Q-Q: All the dots should fall perfectly in line with the red line. Any significant deviations would imply the distribution is skewed.

CORRELOGRAM: Plot show that residual errors are not autocorrelated. In case of  autocorrelation would mean that there is some pattern in the residual errors which are not explained in this model. 

Overall, it seems to be a good fit. Let’s forecast.

```python
# Forecast in a 24 hours horizon
################################

n_periods = 24
fc, confint = model.predict(n_periods=n_periods, return_conf_int=True)
index_of_fc = np.arange(len(poc_16.NO2_index), len(poc_16.NO2_index)+n_periods)

# Plotting
##################################

fc_series = pd.Series(fc, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)


plt.plot(poc_16.NO2_index)
plt.plot(fc_series, color='darkgreen')
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)

plt.title("NO2 index forecast AirStation 16")
plt.show()
```
